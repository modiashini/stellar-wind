```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t_naught = 2.3*10**6
```

```python
r_s = (2*10**6)*(0.181/1)*(t_naught/(2*10**6))**(-1)
```

```python
a_naught = 180*(t_naught/(2*10**6))**(1/2)
```

```python
def stellar_wind(r):
    r *= 10**6
    return a_naught*np.exp(3/2)*np.exp((-2*r_s)/r)
```

```python
x = np.linspace(1,160)
y = stellar_wind(x)
plt.plot(x,y)
plt.title('Stellar Wind of GJ 1132')
plt.xlabel('r')
plt.ylabel('u(r) kms^-1')
plt.show()
```
