```python
#rotation period for M2.5-6.0 V stars
#applies for GJ 1132, GJ 3053, K2-18, GJ 1214
def rotation_period_M2(T):
    return (T-0.012)/0.061

x = np.linspace(1,10)
y = rotation_period_M2(x)

plt.plot(x,y, label="rotation period of M2.5-6.0 V stars")
plt.legend(loc= 'lower right', bbox_to_anchor=(1.29, +0.12))
plt.title('Rotation Period v. Age in Gyr')
plt.xlabel('Age (Gyr)')
plt.ylabel('Rotation Period (days)')
```

```python
#rotation period for M0 V stars
#applies for Kepler 138, K2-3
def rotation_period_M1(T):
    return ((T-0.365)/0.019)**(1/1.457)

x = np.linspace(1,10)
y = rotation_period_M1(x)

plt.plot(x,y, label="rotation period of M0V stars")
plt.legend(loc= 'lower right', bbox_to_anchor=(1.29, +0.12))
plt.title('Rotation Period v. Age in Gyr')
plt.xlabel('Age (Gyr)')
plt.ylabel('Rotation Period (days)')
```

```python
#rotation period for K-type stars
#applies for 55 Cnc, GJ 9827, GJ 97658
def rotation_period_K(T):
    return 25*((T/4.6)**0.5)

x = np.linspace(1,10)
y = rotation_period_K(x)

plt.plot(x,y, label="rotation period of K-Type stars")
plt.legend(loc= 'lower right', bbox_to_anchor=(1.29, +0.12))
plt.title('Rotation Period v. Age in Gyr')
plt.xlabel('Age (Gyr)')
plt.ylabel('Rotation Period (days)')
```
